```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
XTRAIN, XTEST, YTRAIN, YTEST = train_test_split(X,Y)
```

```python
LR = LogisticRegression()
legendary_model = LR.fit(XTRAIN, YTRAIN)
legendary_prediction = legendary_model.predict(XTEST)
accuracy_score(YTEST, legendary_prediction)
```

```python
CM = confusion_matrix(YTEST, legendary_prediction)
CM
```

```python
from sklearn.metrics import roc_curve
```

```python
SY = legendary_model.predict_proba(XTEST)[:,1]
SFPR, STPR, STHRESH = roc_curve(YTEST, SY)
plt.plot(SFPR, STPR, label="sp_attack")
plt.plot([0,1],[0,1],label="is_legendary Guess")
plt.legend()
plt.grid()
plt.xlabel("False Pos")
plt.ylabel("true pos")
plt.title("sp_attack - Legendary Prediction")

plt.gcf().set_size_inches(6,6)
```
